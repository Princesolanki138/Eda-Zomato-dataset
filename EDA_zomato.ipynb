```python
## find the top 10 cusines
top_cuisines = final_df['Cuisines'].str.split(',').explode('Cuisines').str.strip().value_counts().reset_index()[:10]
```

```python
plt.figure(figsize=(12,6)) 
sns.barplot(x="Cuisines", y="count",palette="viridis",hue="Cuisines", data=top_cuisines) 
plt.xlabel("All Top Cuisines")
plt.ylabel("Popularity")
plt.title('Top 10 Cuisines Popularity') 
plt.show()
```

```python
# Total Number Of cuisines each country 
df_exploded = final_df.assign(Cuisines=df['Cuisines'].str.split(',')).explode('Cuisines')
df_exploded['Cuisines'] = df_exploded['Cuisines'].str.strip()


cuisine_country_counts = (
    df_exploded.groupby(['Country', 'Cuisines'])
    .size()
    .reset_index(name='Count')
    .sort_values(['Country', 'Count'], ascending=[True, False])
)


top_10_per_country = cuisine_country_counts.groupby('Country').size().reset_index().rename(columns = {0:"Count"})
top_10_per_country
```

```python
plt.figure(figsize=(20,6)) 
sns.barplot(x="Country", y="Count",palette="viridis",hue="Country", data=top_10_per_country) 
plt.xlabel("Country")
plt.ylabel("Count")
plt.title('Total Number of Cuisines country have ') 
plt.show()
```

India has highest Number of Cuisines 

```python
famous_cuisine_per_country = (
    df_exploded
    .groupby(['Country', 'Cuisines'])
    .size()
    .reset_index(name="Count")
    .sort_values(['Country','Count'], ascending=[True,False])
    .groupby("Country")   # now pick only the top per country
    .head(1)
    .reset_index(drop = True)
)
famous_cuisine_per_country
```

```python
plt.figure(figsize=(12,6))
sns.barplot(
    data=famous_cuisine_per_country,
    x="Country", y="Count", hue="Cuisines", palette="viridis"
)
plt.title("Famous Cuisine in Each Country", fontsize=16)
plt.xticks(rotation=45)
plt.yscale("log") 
plt.ylabel("Count")
plt.xlabel("Country")
plt.legend(title="Cuisine", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```
